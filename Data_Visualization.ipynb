```python
import numpy as np
import pandas as pd
from xlrd import open_workbook
import os.path
df_can = pd.read_excel(r'D:\Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skip_footer=2)
df_can.head()
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
 
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
 
df_can.columns = list(map(str, df_can.columns))
 
df_can.set_index('Country', inplace=True)
 
df_can['Total'] = df_can.sum(axis=1)


# we are using the inline backend
%matplotlib inline 
 
import matplotlib as mpl
import matplotlib.pyplot as plt
 
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
df_can.head()
```

```python
years = list(map(str, range(1980, 2014)))

df_can.loc['Haiti', years].plot(kind='line')
plt.title('Immagiration from Haiti')
plt.xlabel('Years')
plt.ylabel('No.')
plt.show()
```

```python
# Area plot
df_can1 = df_can.head()
df_can1.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_can1.head()
```

```python
df_top5 = df_can1.head()
df_top5 = df_top5[years].transpose()
df_top5.rename(columns={'United Kingdom of Great Britain and Northern Ireland':'UK'}, inplace=True)
df_top5.head()
```

```python
df_top5.plot(kind='area')
plt.title('Immagiration from Haiti')
plt.xlabel('Years')
plt.ylabel('No.')
plt.show()
```
